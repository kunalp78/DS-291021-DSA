## `sort the list using stack`

```python
class Node:
    def __init__(self, val):
        self.data = val
        self.next = None
class Stack:
    def __init__(self):
        self.head = None
        self.size = 0
    
    def isEmpty(self):
        return self.size == 0
    def peek(self):
        if self.isEmpty() is True:
            return "Empty Stack"
        return self.head.data
    def getSize(self):
        return self.size
    
    def push(self, val):
        newNode = Node(val)
        newNode.next = self.head
        self.head = newNode
        self.size+=1
    def pop(self):
        if self.isEmpty() is True:
            return "Stack is Empty"
        else:
            popped_value = self.peek()
            self.head = self.head.next
            self.size-=1
        return popped_value
```

```python
S1 = Stack()
for i in range(ord("A"),ord("F")+1):
    S1.push(chr(i))
```

```python
S1.peek()
```

```python
for _ in range(S1.getSize()):
    print(S1.pop())
```

```python
S1.peek()
```

```python
def sort_stack(stack):
    l = []
    for _ in range(stack.getSize()):
        l.append(stack.pop())
    l.sort(reverse=True)
    for i in l:
        stack.push(i)
        print(stack.peek())
    return stack.peek()
```

# `Queue` using Linked list

```python
Enqueue() --> Adds an item to queue -
                    adds a new node after rear and moves the rear to the next node
Dequeue() --> Removes an item from queue -
                    removes the front node and moves to the next node
```

```python
class Node:
    def __init__(self, val):
        self.data = val
        self.next = None
class Queue:
    def __init__(self):
        self.front = self.rear = None
        self.size = 0
        
    def isEmpty(self):
        return self.front == None
    
    def enQueue(self, data):
        newNode = Node(data)
        
        if self.rear == None:
            self.front = newNode
            self.rear = newNode
            return
        else:
            self.rear.next = newNode
            self.rear = newNode
    def deQueue(self):
        
        if self.isEmpty():
            return "Empty Queue"
        current = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return current
```
